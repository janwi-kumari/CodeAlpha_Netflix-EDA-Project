**Data Cleaning:Handling Missing Values**

```python
df['director']=df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('Unknown')
df['country']=df['country'].fillna('Unknown')
df['date_added']=df['date_added'].fillna('Not Available')
df['rating']=df['rating'].fillna('Not Rated')
df['duration']=df['duration'].fillna('Unknown')
```

```python
df.isnull().sum()
```

**EDA**
Analysis 1: Movies Vs TV Shows

```python
df['type'].value_counts()
```

```python
movie_percent=(6131/8807)*100
tv_percent=(2676/8807)*100

print("Movies Percent:",round(movie_percent,2),'%')
print("TV Percent:",round(tv_percent,2),'%')
```

```python
#Plot
type_count=df['type'].value_counts()
type_count.plot(kind='bar')
plt.title("Movies Vs TV Showws on Netflix")
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

**Observation**
- Movies Type are more on Netflix as comapared to TV Shows
- Movies account for 69.62% while TV Shows holds only 30.38% of Netlix Contents.
- This indicates that Netflix's content library is primarily movie-focused.

Analysis 3: Content Ratings

```python
rating_count=df['rating'].value_counts()
rating_count
```

```python
df=df[df['rating']!='66 min']
df=df[df['rating']!='74 min']
df=df[df['rating']!='84 min']

rating_count=df['rating'].value_counts()
rating_count
```

```python
rating_count.plot(kind='bar')
plt.title("Distribution of Content Ratings")
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.show()
```

**Observations:**

Analysis 4: Top Genres On Netflix

```python
df['listed_in'].value_counts().head(10)
```

```python
genre_list=[]
for genres in df['listed_in']:
    genre_list+=genres.split(',')
genre_list
 
```

```python
genre_series=pd.Series(genre_list)
genre_count=genre_series.value_counts().head(10)
print(genre_count)
```

```python
genre_count.plot(kind='bar')
plt.title('Top 10 Individual Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

The "listed_in" column contains one or more genres for each title. In this i have first worked to get the individual genres from the combination and later have created a bar plot to see the visualizations.

Analysis 9: Top Actors

```python
df['cast'].value_counts()
filtered_cast=df[df['cast']!='Unknown']['cast'].value_counts()
filtered_cast
```

**Data Issues and Limitations**

2. Missing values were replaced with 'Unknown' for analysis purposes.
3. Some records contained incorrect rating values such as '66 min', '74 min','84 min', indicating data quality issues.
4. The 'listed_in' column conatins multiple genres in a single field, making genre analysis less precise unless genres are separated.
5. The dataset represents Netflix titles available at the time of collection and may not reflect the current Netflix catalog.
